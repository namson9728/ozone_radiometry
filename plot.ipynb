```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from the file
data = np.loadtxt('2019Apr30_20_0.out')

# Create the plot
plt.figure(figsize=(27, 18))  # Equivalent to Gnuplot's size 2700x1800
plt.title('SMA - 2019 Apr 30 from 18 UT forecast', fontsize=48, pad=40)
plt.xlabel('Frequency [GHz]', fontsize=48, labelpad=40)
plt.ylabel(r'Zenith dT$_b$', fontsize=48, labelpad=40)

# Set the range for x and y axes
plt.xlim(210, 250)
plt.ylim(25, 75)

# Custom ticks
plt.xticks(np.arange(210, 251, 10), fontsize=36)  # xtics 10
plt.yticks(np.arange(30, 76, 10), fontsize=36)
plt.tick_params(axis='both', which='major', length=20, direction="in", width=4, pad=30)

for spine in plt.gca().spines.values():
    spine.set_linewidth(4)

# Plot the data with lines
plt.plot(data[:, 0], data[:, 1], color='blue', label='20 UT', linewidth=4)

# Add arrows and labels
plt.annotate('HDO', xy=(225.896, 33), xytext=(225.896, 29), ha='center',
             arrowprops=dict(facecolor='gray', shrink=0.05, lw=2), fontsize=30)
plt.annotate('HDO', xy=(241.561, 33), xytext=(241.561, 29), ha='center',
             arrowprops=dict(facecolor='gray', shrink=0.05, lw=2), fontsize=30)
plt.annotate('O^{18}O', xy=(233.946, 33), xytext=(233.946, 29), ha='center',
             arrowprops=dict(facecolor='gray', shrink=0.05, lw=2), fontsize=30)
plt.annotate('O_3', xy=(226.054, 42), xytext=(226.054, 46), ha='center',
             arrowprops=dict(facecolor='gray', shrink=0.05, lw=2), fontsize=30)

# Add legend
plt.legend(loc='upper left', fontsize=36)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = np.loadtxt('2019Apr30_20_0.out')

# Define PWV scale
PWV_SCALE = 2.372  # mm from 2019Apr30_20_0.err

# Calculate dTb/dPWV
frequency = data[:, 0]  # Assuming the first column is frequency
dTb_dPWV = data[:, 3] /PWV_SCALE  # Assuming the fourth column corresponds to dTb

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(frequency, dTb_dPWV, label='20 UT', color='blue')

# Set the y-axis limits and ticks
plt.ylim(9.5, 12.5)  # Set y-axis limits
plt.xlim(210, 250)
plt.yticks(np.arange(10, 13, 1))  # Set y-axis ticks
plt.tick_params(axis='both', which='major', length=10, direction="in", width=0.75, pad=10)

# Set labels and title
plt.ylabel(r'Zenith dT$_b$ / dPWV [K$\cdot$mm$^{-1}$]', fontsize=14, labelpad=20)  # Adjust labelpad for spacing
plt.xlabel('Frequency [GHz]', fontsize=14, labelpad=20)  # Adding x-axis label for clarity
plt.title('SMA - 2019 Apr 30 from  18 UT forecast', fontsize=16, pad=20)

# Add legend
plt.legend(loc='upper left', fontsize=15)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the file (assuming space-separated)
data = np.loadtxt('2019Apr30_20_0.out')

# Extract the relevant columns
x = data[:, 0]  # First column
y = data[:, 5] / data[:, 3]  # 6th column divided by 4th column

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='20 UT', color='blue', linestyle='-', linewidth=1)

# Set y-axis range and ticks
plt.ylim(0.5, 1.0)
plt.yticks(np.arange(0.5, 1.01, 0.1))
plt.xlim(np.min(x), np.max(x))
plt.xticks(np.arange(np.min(x), np.max(x)+1, 10))
plt.tick_params(axis='both', which='major', length=7, direction="in", width=1, pad=10)

# Set axis labels
plt.ylabel(r'zenith (dL / dPWV)(dT$_{\mathrm{sys}}$ / dPWV)$^{-1}$ [mm⋅K$^{-1}$]', size=13)
plt.xlabel('Frequency [GHz]', size=13)

# Set the title and legend
plt.title('2019 Apr 30 - 20 UT Forecast', size=15)
plt.legend(loc="upper left", fontsize=13)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the files (assuming space-separated format)
data_19_0 = np.loadtxt('2019Apr30_19_0.out')
data_19_45 = np.loadtxt('2019Apr30_19_45.out')
data_19_60 = np.loadtxt('2019Apr30_19_60.out')
data_20_0 = np.loadtxt('2019Apr30_20_0.out')
data_20_45 = np.loadtxt('2019Apr30_20_45.out')
data_20_60 = np.loadtxt('2019Apr30_20_60.out')
data_21_0 = np.loadtxt('2019Apr30_21_0.out')
data_21_45 = np.loadtxt('2019Apr30_21_45.out')
data_21_60 = np.loadtxt('2019Apr30_21_60.out')

# Extract relevant columns (1st for x, 6th/4th for y)
def extract_xy(data):
    x = data[:, 0]
    y = data[:, 5] / data[:, 3]
    return x, y

x_19_0, y_19_0 = extract_xy(data_19_0)
x_19_45, y_19_45 = extract_xy(data_19_45)
x_19_60, y_19_60 = extract_xy(data_19_60)

x_20_0, y_20_0 = extract_xy(data_20_0)
x_20_45, y_20_45 = extract_xy(data_20_45)
x_20_60, y_20_60 = extract_xy(data_20_60)

x_21_0, y_21_0 = extract_xy(data_21_0)
x_21_45, y_21_45 = extract_xy(data_21_45)
x_21_60, y_21_60 = extract_xy(data_21_60)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the data with different line styles and labels
plt.plot(x_19_0, y_19_0, label='19 UT', color='gray', linestyle='-', linewidth=1)
plt.plot(x_19_45, y_19_45, color='gray', linestyle='-', linewidth=1)
plt.plot(x_19_60, y_19_60, color='gray', linestyle='-', linewidth=1)

plt.plot(x_20_0, y_20_0, label='20 UT', color='blue', linestyle='-', linewidth=1)
plt.plot(x_20_45, y_20_45, color='blue', linestyle='-', linewidth=1)
plt.plot(x_20_60, y_20_60, color='blue', linestyle='-', linewidth=1)

plt.plot(x_21_0, y_21_0, label='21 UT', color='green', linestyle='-', linewidth=1)
plt.plot(x_21_45, y_21_45, color='green', linestyle='-', linewidth=1)
plt.plot(x_21_60, y_21_60, color='green', linestyle='-', linewidth=1)

# Set y-axis range and ticks
plt.ylim(0.5, 0.95)
plt.yticks(np.arange(0.5, 0.95, 0.1))
plt.xlim(np.min(x), np.max(x))
plt.xticks(np.arange(np.min(x), np.max(x)+1, 10))

# Set axis labels
plt.ylabel(r'zenith (dL / dPWV)(dT$_{\mathrm{sys}}$ / dPWV)$^{-1}$ [mm⋅K$^{-1}$]', size=13)
plt.xlabel('Frequency [GHz]', size=13)

# Add labels for theta_z [deg]
plt.text(220, 0.76, r'$\theta_z$ [deg]', ha='center', fontsize=12)
plt.text(220, 0.73, '60', ha='center', fontsize=12)
plt.text(220, 0.682, '45', ha='center', fontsize=12)
plt.text(220, 0.649, '0', ha='center', fontsize=12)

# Set the title and legend
plt.title('2019 Apr 30 - Zenith dL/dPWV at Different Zenith Angles', size=15)
plt.legend(loc="upper left", fontsize=13)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the file (assuming space-separated format)
data = np.loadtxt('2019Apr30_290K_sim_cal.out')

# Extract relevant columns (1st for x, 2nd for y)
x = data[:, 0]
y = data[:, 1]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the simulated cal load (x vs y)
plt.plot(x, y, label='Simulated cal load', color='blue', linestyle='-', linewidth=1)

# Plot the horizontal line at y=290
plt.axhline(y=290, color='red', linestyle='--', label='Actual cal load temperature')

# Set y-axis range and ticks
plt.ylim(277.5, 292.5)
plt.yticks(np.arange(275, 291, 5))
plt.xlim(np.min(x), 250)
plt.xticks(np.arange(210, 250+0.1, 10))

# Set axis labels
plt.ylabel(r'$T_b$ [K]', size=13)
plt.xlabel('Frequency [GHz]', size=13)

# Add labels for the lines
plt.text(220, 290.5, 'Actual cal load temperature', ha='center', fontsize=12, color='red')
plt.text(220, 288, 'Simulated cal load behind atmosphere', ha='center', fontsize=12, color='blue')

# Set the title and show the plot
plt.title('2019 Apr 30 - Simulated and Actual Cal Load Temperatures', size=15)

# Save the figure
#plt.savefig('2019Apr30_290K_sim_cal.png')

# Show the plot
plt.show()
```
